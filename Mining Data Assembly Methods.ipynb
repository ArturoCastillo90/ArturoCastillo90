```python
import pandas as pd

file_name = "carros_v2.csv"

df = pd.read_csv(file_name)
print("Dimensiones del dataset:", df.shape)

df.head()
```

```python
#Informacin general del dataset
df.info()

#Ver cuntos valores nicos tiene cada columna
print("\nValores nicos por columna:")
for col in df.columns:
    print(f"{col:25s} -> {df[col].nunique()} valores nicos")
```

```python
target_column = "Modelo"
target_column = "Cilindros"
target_column = "Sucursal"
target_column = "Kilometraje"
target_column = "Rendimiento"
target_column = "Precio"
target_column = "Marca"

#Verificar que exista la columna
if target_column not in df.columns:
    raise ValueError(f"La columna '{target_column}' no existe. Revisa los nombres de columnas.")

#Separar X
X = df.drop(columns=[target_column])
y = df[target_column]

print("Shape X:", X.shape)
print("Shape y:", y.shape)
```

```python
# NaNs simples
X = X.fillna("Desconocido")

# Encoding para columnas categricas
X_encoded = pd.get_dummies(X)

print("Shape original X:", X.shape)
print("Shape X codificada:", X_encoded.shape)

X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.3, random_state=42, stratify=y
)

print("Tamao train:", X_train.shape, y_train.shape)
print("Tamao test:", X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report

ada = AdaBoostClassifier(
    n_estimators=100,
    random_state=42
)

ada.fit(X_train, y_train)
ada_pred = ada.predict(X_test)

ada_acc = accuracy_score(y_test, ada_pred)
print(f"Accuracy AdaBoost: {ada_acc:.4f}")

print("\nReporte de clasificacin - AdaBoost:")
print(classification_report(y_test, ada_pred))
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    random_state=42
)

bag.fit(X_train, y_train)
bag_pred = bag.predict(X_test)

bag_acc = accuracy_score(y_test, bag_pred)
print(f"Accuracy Bagging: {bag_acc:.4f}")

print("\nReporte de clasificacin - Bagging:")
print(classification_report(y_test, bag_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

j48 = DecisionTreeClassifier(
    criterion="entropy",   # similar a C4.5
    random_state=42
)

j48.fit(X_train, y_train)
j48_pred = j48.predict(X_test)

j48_acc = accuracy_score(y_test, j48_pred)
print(f"Accuracy J48 (DecisionTree entropy): {j48_acc:.4f}")

print("\n Reporte de clasificacin - J48:")
print(classification_report(y_test, j48_pred))
```

```python
results = {
    "AdaBoost": ada_acc,
    "Bagging": bag_acc,
    "J48": j48_acc
}

print("\n===============================")
print("   COMPARACIN DE MODELOS")
print("===============================")
for model, acc in results.items():
    print(f"{model:10s} -> Accuracy: {acc:.4f}")

```
